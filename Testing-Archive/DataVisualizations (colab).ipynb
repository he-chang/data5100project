```python
# Import pandas, numpy, and matplotlib
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# seaborn is a data visualization library built on matplotlib
import seaborn as sns 

# set the plotting style 
sns.set_style("whitegrid")
```

```python
df = pd.read_csv("https://raw.githubusercontent.com/he-chang/data5100project/main/university_data.csv")
```

```python
df.head()
```

```python
# Set figure size
plt.figure(figsize = (14, 8))

# Setup scatter plot for 
sns.scatterplot(data = df, x = 'co_Asian', y = 'gr_Asian', hue = 'year')

# Label x and y axis
plt.xlabel('Asian Cohort Counts', fontsize = 18)
plt.ylabel('Asian Graduation Count', fontsize = 18)

plt.tick_params(labelsize = 15);
```

```python
# Looking at just year totals
df_gr_year_totals = df.groupby('year')[['gr_total', 'gr_AmerInd_AlaskNtv', 'gr_Asian', 'gr_AfrAmer', 'gr_Hispanic', 'gr_Hi_PacIsland', 'gr_White', 'gr_2orMore', 'gr_nonresAlien', 'co_total', 'co_AmerInd_AlaskNtv', 'co_Asian', 'co_AfrAmer', 'co_Hispanic', 'co_Hi_PacIsland', 'co_White', 'co_2orMore', 'co_nonresAlien']].sum()
df_co_year_totals = df.groupby('year')[['co_total', 'co_AmerInd_AlaskNtv', 'co_Asian', 'co_AfrAmer', 'co_Hispanic', 'co_Hi_PacIsland', 'co_White', 'co_2orMore', 'co_nonresAlien']].sum()
```

```python
# Setup scatter plot for 
df_gr_year_totals.plot(kind = 'barh', alpha = 0.6, figsize = (15, 30))

# Label x and y axis
plt.xlabel('Years', fontsize = 18)
plt.ylabel('Graduation Counts by Ethnicity', fontsize = 18)

plt.tick_params(labelsize = 15);
```

```python
# Setup scatter plot for 
df_co_year_totals.plot(kind = 'barh', alpha = 0.6, figsize = (15, 30))

# Label x and y axis
plt.xlabel('Years', fontsize = 18)
plt.ylabel('Cohort Counts by Ethnicity', fontsize = 18)

plt.tick_params(labelsize = 15);
```
